```python
# Importing necessary libraries
import pandas as pd

# Load the dataset
file_path = 'Iris.csv'
iris_data = pd.read_csv(file_path)

# Show the first few rows of the dataset
iris_data.head()
```

```python
# Importing necessary libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Pairplot to visualize the distribution of data
sns.pairplot(iris_data.drop("Id", axis=1), hue="Species", palette="Set2", diag_kind="kde")
plt.suptitle('Pairplot of Iris Data', y=1.02)
plt.show()

```

```python
# Importing KMeans from sklearn
from sklearn.cluster import KMeans

# Selecting only the numerical features to apply K-means
X = iris_data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

# Applying K-means with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Adding the cluster labels to the original dataframe
iris_data['Cluster'] = kmeans.labels_

# Displaying the first few rows of the dataframe with cluster labels
iris_data.head()
```

# Visualizing the clusters along with the original species labels

```python

plt.figure(figsize=(14, 7))

# Subplot for clusters identified by K-means
plt.subplot(1, 2, 1)
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Cluster', palette='Set1', data=iris_data, legend='full')
plt.title('Clusters Identified by K-means')

# Subplot for original species labels
plt.subplot(1, 2, 2)
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='Set2', data=iris_data, legend='full')
plt.title('Original Species Labels')

plt.suptitle('Comparison of K-means Clusters and Original Species')
plt.tight_layout()
plt.show()
```


Conclusion

Dans cette analyse, nous avons appliqué l'algorithme K-means sur l'ensemble de données Iris, un ensemble de données classique en apprentissage automatique et en statistique. L'ensemble de données comporte quatre caractéristiques mesurables pour trois espèces différentes d'iris : Iris-setosa, Iris-versicolor, et Iris-virginica.

Avant d'appliquer l'algorithme, une analyse exploratoire des données a été réalisée à l'aide de graphiques en paires. Ces graphiques ont suggéré que les différentes espèces sont relativement bien séparées dans l'espace des caractéristiques, ce qui indiquait que l'algorithme K-means pourrait être efficace pour le partitionnement des données en grappes.

Nous avons choisi de partitionner les données en trois grappes (K=3K=3), en adéquation avec le nombre d'espèces dans l'ensemble de données. Les résultats ont montré que les grappes formées par l'algorithme K-means correspondent de manière assez fidèle aux étiquettes d'espèces originales. Une visualisation des grappes et des étiquettes d'espèces sur les caractéristiques "SepalLengthCm" et "SepalWidthCm" a confirmé cette correspondance.

Cependant, il est important de noter que quelques points ont été classés différemment par l'algorithme K-means par rapport à leurs étiquettes d'espèces originales. Cela est attendu car K-means est un algorithme non supervisé et n'avait pas accès aux étiquettes d'espèces pendant l'apprentissage.

En somme, cette analyse démontre l'efficacité de l'algorithme K-means pour le partitionnement de données quand une bonne séparation existe entre les grappes. Cela suggère que l'algorithme peut être utile dans des applications pratiques où l'étiquetage manuel des données est difficile ou coûteux.
