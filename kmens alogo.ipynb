```python
# Importing necessary libraries
import pandas as pd

# Load the dataset
file_path = 'Iris.csv'
iris_data = pd.read_csv(file_path)

# Show the first few rows of the dataset
iris_data.head()
```

```python
# Importing necessary libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Pairplot to visualize the distribution of data
sns.pairplot(iris_data.drop("Id", axis=1), hue="Species", palette="Set2", diag_kind="kde")
plt.suptitle('Pairplot of Iris Data', y=1.02)
plt.show()

```

```python
# Importing KMeans from sklearn
from sklearn.cluster import KMeans

# Selecting only the numerical features to apply K-means
X = iris_data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

# Applying K-means with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Adding the cluster labels to the original dataframe
iris_data['Cluster'] = kmeans.labels_

# Displaying the first few rows of the dataframe with cluster labels
iris_data.head()
```

# Visualizing the clusters along with the original species labels

```python

plt.figure(figsize=(14, 7))

# Subplot for clusters identified by K-means
plt.subplot(1, 2, 1)
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Cluster', palette='Set1', data=iris_data, legend='full')
plt.title('Clusters Identified by K-means')

# Subplot for original species labels
plt.subplot(1, 2, 2)
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='Set2', data=iris_data, legend='full')
plt.title('Original Species Labels')

plt.suptitle('Comparison of K-means Clusters and Original Species')
plt.tight_layout()
plt.show()
```
